```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

df = pd.read_csv('./titanic.csv', sep=',')
df.head()
```

```python
#생존자 명수 - 1:생존 0: 사망
df['Survived'].value_counts()
```

```python
# 항구 'S'에서 탑승한 승객수
df['Embarked'].value_counts()
#df[(df['Embarked'] == 'S')]
```

```python
#남성 승객수
df['Sex'].value_counts()
```

```python
#3등실에 탑승하고, 지불한 요금이 40 미만인 승객 수
df[(df['Pclass'] == 3) & (df['Fare'] < 40.0)].count()
```

```python
# 승객 ID 674인 승객 이름
condition1 = (df['PassengerId'] == 674)
df.loc[condition1, ['Name']].head()
```

```python
# 승객들의 평균 나이
df.describe()
```

```python
#생존자와 사망자의 수
df['Survived'].value_counts()
```

```python
#남성 승객수
df[(df['Sex'] == 'male')]
```

```python
#여성 승객수
df[(df['Sex'] == 'female')]
```

```python
#각 성별에 따른 생존자와 사망자 수 
df.groupby(['Sex','Survived'])['Survived'].count()
```

```python
#각 등급별로 생존한 사람과 사망한 사람의 평균 지불 요금
df.groupby(['Pclass','Sex','Survived'])['Fare'].mean()
```

```python
#탑승자의 생존 여부에 대한 비율을 계산하고, 이 비율을 막대 그래프
sur = df['Survived'].value_counts().plot.bar()
sur.set_title("생존률 \n")
```

```python
#나이 열의 분포를 20개의 구간(bin)으로 나누어 히스토그램으로 시각화

data = df['Age']
plt.hist(data, bins=20)  # 데이터를 20개의 구간으로 나눔
plt.xlabel('Value')
plt.ylabel('Age')
plt.title('')
plt.show()
```

```python
#나이와 요금 간의 산점도 시각화
df.plot(kind='scatter', x='Age', y='Fare')
```

```python
#객실 등급(Pclass)별로 생존자와 사망자 수를 막대그래프로 시각화
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')

sns.countplot(data=df, x='Pclass', hue='Survived', )
plt.show()
```
